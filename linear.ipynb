# T-Test

**Definition:** T-Test is used to compare means when the population variance is unknown or the sample size is small.

---

**Hypotheses:**  
- H0: distrubution of tip for yellow= distrubution of tip for green 
- H1: distrubution of tip for yellowâ‰  distrubution of tip for green


```python
yellow_tip=df['tip'][df['color']=='yellow']
green_tip=df['tip'][df['color']=='green']
```

```python
yellow_tip
```

```python
green_tip
```

```python
green_tip.shape
```

```python
yellow_tip.shape
```

```python
from scipy import stats
t_stat, p_val = stats.ttest_ind(yellow_tip,green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python

if p_val < 0.05:
    print("Reject null hypothesis:The tip for yellow taxis is different that of green")
else:
    print("Fail to reject null hypothesis")
```

** Hypotheses: **
- H0: distribution tip is the same across all pickup_borogh categories
- H1: At least one region has a different mean spending score

```python
manhattan_tip=df['tip'][df['pickup_borough']=='Manhattan']
Queens_tip=df['tip'][df['pickup_borough']=='Queens']
Bronx_tip=df['tip'][df['pickup_borough']=='Bronx']
Brooklyn_tip=df['tip'][df['pickup_borough']=='Brooklyn']
```

```python
f_stat, p_val = stats.f_oneway(manhattan_tip, Queens_tip, Bronx_tip, Brooklyn_tip)
#print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)
alpha=0.05
if p_val < alpha:
    print("Reject null hypothesis: ")
else:
    print("Fail to reject null hypothesis: ")
```

Hypotheses for Tukey test (for each pair of groups):
. HO: distribution of group1 = distribution of group2
. H1: distribution of group1 # distribution of group2

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
# Perform Tukey HSD
tukey_result=pairwise_tukeyhsd(endog=df.tip,groups=df.pickup_borough,alpha=0.05)
print(tukey_result)
```

2 tests we have pearson and sperman r test both test are used to analyze two numberical columns not for the sample but population,we give some assumpation base don the population of the  sample

Hypotheses for correlation for numeric variables:
Â· HO: correlation is zero
Â· H1: there is significant correlation (not zero)

```python
r, p_val=stats.pearsonr(df['tip' ],df['passengers' ])
print("correlation", r)
print("P-value:", p_val)
alpha=0.05
if p_val < alpha:
    print("Reject null hypothesis: ")
else:
    print ("Fail to reject null hypothesis: ")
```

```python
r, p_val=stats.spearmanr(df['tip' ],df['passengers' ])
print("correlation", r)
print("P-value:", p_val)
alpha=0.05
if p_val < alpha:
    print("Reject null hypothesis:There is no relationship ")
else:
    print ("Fail to reject null hypothesis: There is significant relationship")
```
